#### Importing libraries

```python
import pandas as pd
import numpy as np
```

#### Importing the dataset

```python
data = pd.read_csv('osmi2014.csv')
```

#### Exploring the data

```python
data.head()
```

```python
# age is numerical, every other is categorical
data.dtypes
```

```python
# finding the null values
data.isnull().sum()
```

#### Drop irrelevant columns

```python
data.drop(columns=['state','comments', 'Timestamp'], inplace=True)
```

##### Rename columns 

```python
data.rename(columns = {'Age': 'age'}, inplace=True)
data.rename(columns = {'Gender': 'gender'}, inplace=True)
data.rename(columns = {'Country': 'country'}, inplace=True)
```

```python
data.head()
```

##### Impute NaN (self_employed = 18, work_interfere = 264) 

```python
print(data['self_employed'].value_counts())
print(data['work_interfere'].value_counts())
```

```python
data['self_employed'].fillna('No', inplace=True)
data['work_interfere'].fillna('Never', inplace=True)
```

```python
data.head()
```

##### Correction of Outliers

```python
data['age'].unique()
```

```python
# drop 'age' outliers
data.drop(data[data['age'] < 18].index, inplace = True) 
data.drop(data[data['age'] > 99].index, inplace = True) 
```

```python
data['age'].unique()
```

```python
data['gender'].unique()
```

```python
# drop 'gender' outliers
data['gender'].replace(['M', 'Male', 'male', 'm', 'Male-ish', 'maile', 'Cis Male', 'Mal', 'Male (CIS)', 'Make', 'Male ', 'Man', 'msle', 'Mail', 'cis male', 'Malr', 'Cis Man'], 'Male', inplace=True)
data['gender'].replace(['Female', 'female', 'Cis Female', 'F', 'Woman', 'f', 'Femake', 'woman', 'Female ', 'cis-female/femme', 'Female (cis)', 'femail'], 'Female', inplace=True)
data['gender'].replace(['Trans-female', 'something kinda male?', 'queer/she/they', 'non-binary', 'Nah', 'Enby', 'fluid',
       'Genderqueer', 'Androgyne', 'Agender', 'Guy (-ish) ^_^', 'male leaning androgynous', 'Trans woman', 'Neuter', 'Female (trans)', 'queer', 'ostensibly male, unsure what that really means'], 'Other', inplace=True)
```
